```python
import json
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Open the file for reading
with open('y_pred_list.json', 'r') as F:
    y_pred = json.loads(F.read())
with open('y_true_list.json', 'r') as F:
    y_true = json.loads(F.read())
```

```python
len(y_pred)
```

```python
sns.set_theme()
plt.rcParams.update({'font.size': 22})
fig, axs = plt.subplots(4,2, figsize=(15,15), sharey=False)


model_label = 'model'
truth_label = 'ground truth'


axs[0,0].plot((np.array(y_pred[0])>.5).astype(int).cumsum(), label= model_label)
axs[0,0].plot(np.array(y_true[0]).cumsum(), label=truth_label)
axs[0,0].text(0.1,0.9,'S05E06',
    horizontalalignment='center',
    verticalalignment='center',
    transform = axs[0,0].transAxes)

axs[0,1].plot((np.array(y_pred[1])>.5).astype(int).cumsum(), label=model_label)
axs[0,1].plot(np.array(y_true[1]).cumsum(), label=truth_label)
axs[0,1].text(0.1,0.9,'S05E08',
    horizontalalignment='center',
    verticalalignment='center',
    transform = axs[0,1].transAxes)

axs[1,0].plot((np.array(y_pred[2])>.5).astype(int).cumsum(), label=model_label)
axs[1,0].plot(np.array(y_true[2]).cumsum(), label=truth_label)
axs[1,0].text(0.1,0.9,'S05E10',
    horizontalalignment='center',
    verticalalignment='center',
    transform = axs[1,0].transAxes)

axs[1,1].plot((np.array(y_pred[3])>.5).astype(int).cumsum(), label=model_label)
axs[1,1].plot(np.array(y_true[3]).cumsum(), label=truth_label)
axs[1,1].text(0.1,0.9,'S05E12',
    horizontalalignment='center',
    verticalalignment='center',
    transform = axs[1,1].transAxes)

axs[2,0].plot((np.array(y_pred[4])>.5).astype(int).cumsum(), label=model_label)
axs[2,0].plot(np.array(y_true[4]).cumsum(), label=truth_label)
axs[2,0].text(0.1,0.9,'S05E13',
    horizontalalignment='center',
    verticalalignment='center',
    transform = axs[2,0].transAxes)

axs[2,1].plot((np.array(y_pred[5])>.5).astype(int).cumsum(), label=model_label)
axs[2,1].plot(np.array(y_true[5]).cumsum(), label=truth_label)
axs[2,1].text(0.1,0.9,'S05E17',
    horizontalalignment='center',
    verticalalignment='center',
    transform = axs[2,1].transAxes)

axs[3,0].plot((np.array(y_pred[6])>.5).astype(int).cumsum(), label=model_label)
axs[3,0].plot(np.array(y_true[6]).cumsum(), label=truth_label)
axs[3,0].text(0.1,0.9,'S05E21',
    horizontalalignment='center',
    verticalalignment='center',
    transform = axs[3,0].transAxes)

axs[3,1].plot((np.array(y_pred[7])>.5).astype(int).cumsum(), label=model_label)
axs[3,1].plot(np.array(y_true[7]).cumsum(), label=truth_label)
axs[3,1].text(0.1,0.9,'S05E22',
    horizontalalignment='center',
    verticalalignment='center',
    transform = axs[3,1].transAxes)
    
fig.tight_layout()
lines_labels = [axs[3,1].get_legend_handles_labels()]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.legend(lines, labels, loc='upper center', ncol=2, bbox_to_anchor=(0.5, 0), bbox_transform=plt.gcf().transFigure)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = confusion_matrix((np.array(y_pred[2])>.5).astype(int),np.array(y_true[2]), normalize='all')

disp = ConfusionMatrixDisplay(cm, display_labels=['Not Perp', 'Perp'])
disp.plot()
```

```python
tot_cm = np.zeros((2,2))
for y_h, y in zip(y_pred, y_true):
    cm = confusion_matrix(np.array(y),(np.array(y_h)>.5).astype(int), normalize='all')
    tot_cm += cm
```

```python
disp = ConfusionMatrixDisplay(tot_cm, display_labels=['Not Perp', 'Perp'])
disp.plot()
```

```python
y_t = []
for y in y_true:
    y_t += y

y_h = []
for y in y_pred:
    y_h += y

font = {'family' : 'normal',
    'weight' : 'bold',
    'size'   : 45}
plt.rc('font', **font)

fig, ax = plt.subplots(figsize=(20, 20))
# cm = confusion_matrix(np.array(y_t),(np.array(y_h)>.5).astype(int), normalize='all')

disp = ConfusionMatrixDisplay.from_predictions(
    np.array(y_t),
    (np.array(y_h)>.5).astype(int), 
    display_labels=['Not Perp', 'Perp'],
    normalize='all',
    ax=ax)
```
